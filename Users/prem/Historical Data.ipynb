```python
os.chdir(r'G:\project\data_03052019\Banks - Private Sector')

d = {}
for i, j in enumerate(os.listdir('.')):
    df_name = j.rstrip('.csv').replace(' ','')
    d[df_name] = pd.read_csv(j, header = None)
    d[df_name]['Stock'] = j.rstrip('.csv')
    d[df_name].columns = ['date', 'open', 'high', 'low', 'close', 'volume', 'Bonus', 'dividend', 'Rights : Ratio', 'split', 'Stock' ]
```

```python
result_list = []

for i in d.keys():
    result_list.append((  (d[i]['date'].head(1).iloc[0], d[i]['Stock'].head(1).iloc[0])))
    
for i, j in sorted(result_list):
    print(j , '===' , i.strftime('%d-%b-%Y'))
```

```python
AllTimeLow_dict = {}

AllTimeLow_dict['stock'] = []
AllTimeLow_dict['date'] = []
AllTimeLow_dict['low_value'] = []

for i in d.keys():
    
    index_pos = d[i]['low'].idxmin()
    AllTimeLow_dict['stock'].append( d[i].iloc[index_pos]['Stock'])
    AllTimeLow_dict['date'].append( d[i].iloc[index_pos]['date'])
    AllTimeLow_dict['low_value'].append( d[i].iloc[index_pos]['low'])
```

```python
AllTimeLow = pd.DataFrame.from_dict(AllTimeLow_dict)
```

```python
AllTimeHigh_dict = {}

AllTimeHigh_dict['stock'] = []
AllTimeHigh_dict['date'] = []
AllTimeHigh_dict['high_value'] = []

for i in d.keys():
    
    index_pos = d[i]['high'].idxmax()
    AllTimeHigh_dict['stock'].append( d[i].iloc[index_pos]['Stock'])
    AllTimeHigh_dict['date'].append( d[i].iloc[index_pos]['date'])
    AllTimeHigh_dict['high_value'].append( d[i].iloc[index_pos]['high'])
```

```python
AllTimehigh = pd.DataFrame.from_dict(AllTimeHigh_dict)
```

```python
merged_df = pd.merge(AllTimeLow, AllTimehigh, on='stock', how='inner')
```

```python
merged_df.rename({'date_x' : 'AllTimeLow', 'date_y' : 'AllTimeHigh'}, axis=1, inplace=True)
```

```python
merged_df
```

```python
merged_df['max_change'] = (merged_df['high_value'] - merged_df['low_value']) / merged_df['low_value'] * 100
```

```python
merged_df['duration'] = (merged_df['AllTimeHigh'].dt.year - merged_df['AllTimeLow'].dt.year).apply(str) + ' Yrs ' + ((merged_df['AllTimeHigh'].dt.month - merged_df['AllTimeLow'].dt.month)% 12).apply(str) + ' Months'
```

```python
merged_df.sort_values('max_change', ascending=False)
#duration is the time taken to achieve it's all time high from all time low
```

clearly says, that Kotak Mahindra Bank has seen the maximum up's and downs, it had all time high just a few days ago which is nearly 1100 times to it's all time low followed by IndusInd Bank, both share the common date for their all time low

There are three banks for which the duration is low, could say the are heading in the negative way


```python
result_dict = {}

from dateutil.relativedelta import relativedelta

no_of_days = 365

for i in d.keys():
    start_date = datetime.datetime.strptime('01 Apr 2005', '%d %b %Y')
    
    if not (d[i].tail(1)['date'].iloc[0] - d[i].head(1)['date'].iloc[0]).days < (datetime.datetime.now() - start_date).days:
        return_list = []
        
        while start_date < datetime.datetime.strptime('01 Apr 2019', '%d %b %Y'):
            end_date = start_date + relativedelta(years=1, days =-1)
            open_value = d[i][d[i]['date'] == start_date]['close']
            close_value = d[i][d[i]['date'] == end_date]['close']            
    
            while open_value.empty:
                start_date = start_date + relativedelta(days =+1)
                open_value = d[i][d[i]['date'] == start_date]['close']

            while close_value.empty:
                end_date = end_date + relativedelta(days =-1)
                close_value = d[i][d[i]['date'] == end_date]['close']
                
            return_list.append(((close_value.iloc[0] - open_value.iloc[0]) / open_value.iloc[0] * 100).round(2))
            
    
            start_date = (start_date + relativedelta(years =+1)).replace(month=4, day=1)
        
        else:
            result_dict[i] = return_list
```

```python
result_df = pd.DataFrame(result_dict, index=range(2006, 2020))
```
